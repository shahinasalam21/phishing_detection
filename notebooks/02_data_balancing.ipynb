```python
import pandas as pd

# Load extracted features
df = pd.read_csv("../data/processed/urls_with_features.csv")

print(df.shape)
print(df['label_numeric'].value_counts())
```

```python
df_legit = df[df['label_numeric'] == 0]
df_phish = df[df['label_numeric'] == 1]

print("Legitimate:", df_legit.shape)
print("Phishing:", df_phish.shape)

```

```python
df_legit_downsampled = df_legit.sample(
    n=len(df_phish),
    random_state=42
)
```

```python
df_balanced = pd.concat([df_legit_downsampled, df_phish])

df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
print(df_balanced.shape)
print(df_balanced['label_numeric'].value_counts())
```

```python
df_balanced.to_csv("../data/processed/balanced_urls_with_features.csv", index=False)

print("Balanced dataset saved successfully")
```
